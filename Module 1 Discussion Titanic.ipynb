```python
import pandas as pd
import numpy as np
df = pd.read_csv('train.csv')

print(df.describe())
```

```python
import matplotlib.pyplot as plt

labels = 'Survived', 'Not survived'
C1 = [C1S,100-C1S]
C2 = [C2S,100-C2S]
C3 = [C3S,100-C3S]
colors = ["#4e2a84", "#ccc4df"]

fig = plt.figure(figsize=(15,10))

plt.subplot(1, 3, 1)
plt.pie(C1, colors=colors, labels=labels, autopct='%1.1f%%', textprops={'size': 'larger', 'color':'black'})
plt.title('Class 1 Survival')
circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)

plt.subplot(1, 3, 2)
plt.pie(C2, colors=colors, labels=labels, autopct='%1.1f%%', textprops={'size': 'larger', 'color':'black'})
plt.title('Class 2 Survival')
circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)

plt.subplot(1, 3, 3)
plt.pie(C3, colors=colors, labels=labels, autopct='%1.1f%%', textprops={'size': 'larger', 'color':'black'})
plt.title('Class 3 Survival')
circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)

plt.show()
```

```python
C1S = (len(df[(df['Survived']==1) & (df['Pclass']==1)])/len(df[(df['Pclass']==1)]))*100
         
C2S = (len(df[(df['Survived']==1) & (df['Pclass']==2)])/len(df[(df['Pclass']==2)]))*100
            
C3S = (len(df[(df['Survived']==1) & (df['Pclass']==3)])/len(df[(df['Pclass']==3)]))*100

print(C1S,C2S,C3S)
```

```python
N = 3
Srv = (C1S,C2S,C3S)
NSrv = (100-C1S,100-C2S,100-C3S)

ind = np.arange(N) 
width = 0.35

p1 = plt.bar(ind, Srv, width, color="#4e2a84")
p2 = plt.bar(ind, NSrv, width, color="#ccc4df", bottom=Srv)

plt.ylabel('Percent Survived')
plt.title('Survival Rate by Class')
plt.xticks(ind, ('1st Class','2nd Class','3rd Class'))
plt.ylim(0,100)

plt.legend((p1[0], p2[0]), ('Survived', 'Not Survived'),loc=1)


#plt.show()
plt.savefig('Survival Rate by Class.pdf')
```
